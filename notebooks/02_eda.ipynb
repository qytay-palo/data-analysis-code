```python
import matplotlib.pyplot as plt
import seaborn as sns

def bar_graph(df:pl.DataFrame, x: str, y: str, hue : str= None, title: str = None):
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df.to_pandas(), x=x, y=y, hue=hue)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
bar_graph(workforce_df, "sector", "count", None, "Workforce Distribution by Sector")
```
